1. Develop a function that takes input: (String: planet, Float: weight in pounds) of object on Earth.  Returns float: kilos of object on that planet

```python
# Define a dictionary data
data = {"Sun" : 27.01, "Mercury" : 0.38,
        "Venus" : 0.91, "Moon" : 0.166,
        "Mars" : 0.38, "Jupiter" : 2.34,
        "Saturn" : 1.06, "Uranus" : 0.92,
        "Neptune" : 1.19, "Pluto" : 0.6}

# A function that returns weight in kilos
def my_func(planet:str, number:float):
  for key, value in data.items():
    try:
      number = float(number)                      # convert number to float
      planet = planet.capitalize()                # change planet to capitalization
      if key == planet and value == number:
        # 1 pound = 0.453592 kilogram
        number_kilos = number*0.453592
        break
      elif key == planet and value != number:
        number_kilos = value*0.453592
        break
      elif key != planet and value == number:
        number_kilos = number*0.453592
        planet = key
        break
        print("Planet not found")
    except ValueError:
      return "Invalid input: reqiures integer or float"
  return f"Planet: {planet} Kilos: {number_kilos}"
```

Display at least 3 tests with various arguments.

```python
print(my_func("sun", 27.01))
print("----------------------")
print(my_func("Moon", 0.166))
print("----------------------")
print(my_func("Neptune", 1.19))
print("----------------------")
print("Try with input error output")
print(my_func("Nep", 27.01))
print("----------------------")
print(my_func("Neptune", "bhhhg"))
print("----------------------")
```

2. Create a class and objects as described

```python
# Create the product class
class Product:
    """

    Attributes:
      name:
      price:
      quantity:
    """
    # A constructor that initializes the attributes
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    # Increase the quantity of the product by the given amount
    def restock(self, amount):
        if amount > 0:
            self.quantity += amount
            print(f"Restock {amount} New {self.name}: {self.quantity} left")
        else:
            print("Restock amount must be positive.")
    # Decrease the quantity by the given amount if there is enough stock
    def sell(self, amount):
        if amount > self.quantity:
            print("Insufficient stock.")
        elif amount <= 0:
            print("sell amount must be positive.")
        else:
            self.quantity -= amount
            print(f"Sold {amount} New {self.name}s, {self.quantity} left")

    def get_quantity(self):
        return self.quantity

    # return a string representation of the product, including its name, price, and quantity.
    def __str__(self):
        return f"Product Name: {self.name}\nPrice: ${self.price}\nQuantity: {self.quantity}"


product_1 = Product("Laptop", 1200, 10)
print(f"Product 1\n{product_1}")
product_1.sell(3)
print("-----------------------")
product_2 = Product("Smartphone", 800, 25)
print(f"Product 2\n{product_2}")
product_2.sell(5)
print("-----------------------")
product_3 = Product("Headphones", 150, 50)
print(f"Product 3\n{product_3}")
product_3.restock(30)
```

6. Using loc, find out which employees belong to the department ‘101’

```python
# Create a dictionary with a variable name, employees
employees = {
    'EmployeeID': [1, 2, 3, 4, 5],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'DepartmentID': [101, 102, 101, 103, 102]}

# Load employees dictionary data with pandas
df_employees = pd.DataFrame(employees)
```

```python
#  find out which employees belong to the department ‘101’ approach 1
data_loc = df_employees.loc[df_employees["DepartmentID"] == 101]
print(f"filtering with loc:\n{data_loc}")
print("---------------------------------------------------------")

# Approach 2
data = df_employees[df_employees["DepartmentID"] == 101]
print(f"filtering with an operator:\n{data}")
print("---------------------------------------------------------")
```
